```python
# Creating a dataframe with USA 2020 events only
df2 = df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.head()
```

```python
df2.shape
```

```python
# Remove USA from event name
df2['Event name'].str.split('(').str.get(0)
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
# get athelete age
df2['athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
# Format athlete performance column
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head()
```

```python
# Drop unnecessary columns (Athlete club, Athlete Country, Athlete Year of birth, Athlete Age category)
df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis = 1)
```

```python
df2.head()
```

```python
# Remove Null Values
df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna() == 1]
```

```python
df2 = df2.dropna()
```

```python
df2.head()
```

```python
# Remove duplicate values if exists
df2[df2.duplicated() == True]
```

```python
# Reset index
df2.reset_index(drop = True)
```

```python
# Fix Dataypes
df2.dtypes
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int)
```

```python
df2.dtypes
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
# Rename columns
df2 = df2.rename(columns = {'Year of event': 'year',
                 'Event dates': 'race_date',
                 'Event name': 'race_name',
                 'Event distance/length': 'race_length' ,
                 'Event number of finishers': 'race_number_of_finisher',
                 'Athlete performance': 'athlete_performance',
                 'Athlete gender': 'athlete_gender',
                 'Athlete average speed': 'athlete_avg_speed',
                 'Athlete ID': 'athlete_ID' 
})
```

<h2>Finding Insights from data</h2>

```python
df2.info()
```

```python
#Compare avg speed of male and female athlete in different race types
df2.groupby(['race_length', 'athlete_gender'])['athlete_avg_speed'].mean()
```

```python
# Find the age group that performs best in 50mi
df2.query('race_length == "50mi"').groupby('athlete_age')['athlete_avg_speed'].agg(['mean','count']).sort_values('mean', ascending = False).head(15)
```

```python
# Find the age group that performs worst in 50mi
df2.query('race_length == "50mi"').groupby('athlete_age')['athlete_avg_speed'].agg(['mean','count']).sort_values('mean', ascending = True).head(15)
```

```python
# Find how seasons affect the speed of athlete

# winter -> 12-2
# spring -> 3-5
# summer -> 6-8
# fall -> 9-11
```

```python
df2['race_month'] = df2['race_date'].str.split('.').str.get(1).astype(int)
```

```python
df2.head()
```

```python
df2['race_season'] = df2['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
df2.head(10)
```

```python
df2.tail(10)
```

```python
df2.groupby('race_season')['athlete_avg_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```

```python
# Comapre avg speed for 50 miles only
df2.query('race_length == "50mi"').groupby('race_season')['athlete_avg_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```

```python
# Comapre avg speed for 50 km only
df2.query('race_length == "50"').groupby('race_season')['athlete_avg_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```
